```python
#What’s the average amount per order in each country?
import pandas as pd
import numpy as np
```

```python
#load data file
DataFlie = pd.read_csv(r"D:\DataSET\practics_Cosmetics_&_Skincare_Product_Sales_Data_(2022)\cosmetics_sales_data.csv")
DataFlie.head(5)
```

```python
dataforDA = DataFlie[['Country','Amount ($)','Boxes Shipped']]
dataforDA 
```

```python
sum = dataforDA.groupby('Country')[['Amount ($)', 'Boxes Shipped']].sum().reset_index()
sum
```

```python
sum['Average'] = sum['Amount ($)']/sum['Boxes Shipped']
averageamountperorder = sum
averageamountperorder
```

```python
import matplotlib.pyplot as plt

# Sort by average for cleaner visuals (optional)

# Plot
plt.figure(figsize=(12, 6))
plt.bar(averageamountperorder['Country'], averageamountperorder['Average'], color='skyblue')

# Annotate
plt.xlabel('Country')
plt.ylabel('Average Amount Per Order')
plt.title('Average Order Value by Country')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
